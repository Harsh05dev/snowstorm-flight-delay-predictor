## Lesson 5: Filter rows (THIS IS HUGE)
This is how you say "show me only flights where ___"

Think of it as a search filter on a website.

```python
# "Show me only flights where it was snowing"
# Step 1: df['SNOW'] > 0 creates True/False for each row

df['SNOW'] > 0

# True  = this flight had snow
# False = no snow
```

```python
# Step 2: Put that True/False inside df[] to FILTER

snowy_flights = df[df['SNOW'] > 0]

print(f'Total flights:  {df.shape[0]}')
print(f'Snowy flights:  {snowy_flights.shape[0]}')

# Read it as: "from df, give me only rows WHERE snow is greater than 0"
```

```python
# More examples:

# Flights where temp was below freezing
cold_flights = df[df['TMAX'] < 32]
print(f'Below freezing flights: {cold_flights.shape[0]}')

# Flights on Delta
delta_flights = df[df['CARRIER_NAME'] == 'Delta Air Lines Inc.']
print(f'Delta flights: {delta_flights.shape[0]}')

# Flights that were delayed
delayed_flights = df[df['DEP_DEL15'] == 1]
print(f'Delayed flights: {delayed_flights.shape[0]}')
```

```python
# Combine filters with & (AND)
# "Flights where it was snowing AND the flight was delayed"

snowy_and_delayed = df[(df['SNOW'] > 0) & (df['DEP_DEL15'] == 1)]
print(f'Snowy AND delayed: {snowy_and_delayed.shape[0]}')

# IMPORTANT: each condition needs its own parentheses ()
```

## Lesson 7: Create a new column
This is the foundation of feature engineering!

```python
# Create a new column that says True/False if it's below freezing

df['IS_FREEZING'] = df['TMAX'] < 32

# That's it! You just created a new column.
# Let's check it:
df[['TMAX', 'IS_FREEZING']].head(10)
```

```python
# Make it 0 and 1 instead of True/False (ML models prefer numbers)

df['IS_FREEZING'] = (df['TMAX'] < 32).astype(int)

# .astype(int) converts True→1, False→0
df[['TMAX', 'IS_FREEZING']].head(10)
```

```python
# You can also do math to create columns

# Convert temp from Fahrenheit to Celsius
df['TEMP_CELSIUS'] = (df['TMAX'] - 32) * 5/9

df[['TMAX', 'TEMP_CELSIUS']].head()
```

## Lesson 8: groupby() — Analyze patterns
"What's the average delay rate for each airline?"

```python
# Average delay rate per month
# groupby('MONTH') = group all flights by their month
# ['DEP_DEL15'].mean() = for each group, what % were delayed?

df.groupby('MONTH')['DEP_DEL15'].mean()
```

```python
# Average delay rate per airline
df.groupby('CARRIER_NAME')['DEP_DEL15'].mean().sort_values(ascending=False)
```

```python
# Average snowfall per month
df.groupby('MONTH')['SNOW'].mean()
```

## Lesson 9: Simple charts
Visualize what you just learned.

```python
import matplotlib.pyplot as plt

# Bar chart: delay rate by month
df.groupby('MONTH')['DEP_DEL15'].mean().plot(kind='bar')
plt.title('Delay Rate by Month')
plt.ylabel('Delay Rate')
plt.show()
```

```python
# Histogram: distribution of snowfall
df['SNOW'].plot(kind='hist', bins=30)
plt.title('Snowfall Distribution')
plt.xlabel('Snowfall (inches)')
plt.show()
```

```python
# Horizontal bar: flights per airline (top 8)
df['CARRIER_NAME'].value_counts().head(8).plot(kind='barh')
plt.title('Top 8 Airlines by Flight Count')
plt.xlabel('Number of Flights')
plt.show()
```

## What you now know:

| What | How | Example |
|------|-----|---------|
| Load data | `pd.read_csv()` | `df = pd.read_csv('file.csv')` |
| See data | `.head()`, `.shape` | `df.head()`, `df.shape` |
| One column | `df['col']` | `df['SNOW']` |
| Many columns | `df[['a','b']]` | `df[['SNOW','TMAX']]` |
| Filter rows | `df[df['col'] > x]` | `df[df['SNOW'] > 1]` |
| Count | `.value_counts()` | `df['MONTH'].value_counts()` |
| New column | `df['new'] = ...` | `df['COLD'] = (df['TMAX'] < 32).astype(int)` |
| Group stats | `.groupby()` | `df.groupby('MONTH')['SNOW'].mean()` |
| Charts | `.plot(kind='bar')` | `data.plot(kind='bar')` |
